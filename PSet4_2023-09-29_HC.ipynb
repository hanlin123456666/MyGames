A. Scatter Plot: visualize the relationship between two numerical variables. Label the axes appropriately and provide a title for the plot.

```python
# Generative AI foundation
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(10, 6))
sns.scatterplot( x='TOTAL_VACCINATIONS', y='PERSONS_LAST_DOSE',data=df)
```

```python
# Manual aesthetic improvements
plt.xlabel('Total Vaccinations',fontsize=15)
plt.ylabel('Persons of Last Dose (billion)',fontsize=15)
plt.title('Relationship between Total Vaccinations and Persons of Last Dose',fontsize=15)
sns.scatterplot(data=df,x='TOTAL_VACCINATIONS', y='PERSONS_LAST_DOSE',hue='PERSONS_LAST_DOSE',s=100)
top_two = df.nlargest(2, 'PERSONS_LAST_DOSE')
for i, point in top_two.iterrows():
    plt.text(point['TOTAL_VACCINATIONS'], point['PERSONS_LAST_DOSE'], point['PERSONS_LAST_DOSE']/1000000000, fontsize=10, verticalalignment='bottom', horizontalalignment='right')
plt.show()

# Alt-text description
'''This plot is about the relationship between total vaccinations and 
persons of last dose. The x-axis shows total vaccinations, which means cumulative 
total vaccine doses administered. The y-axis shows persons of last dose, which is 
the cumulative number of persons vaccinated with a complete primary series. 
Persons vaccinated with a complete primary series refer to people who have 
received all required doses of a COVID-19 vaccine to be considered fully 
vaccinated. The figure demonstrates as the number of
total vaccinations increases, the number of persons of last dose increases.'''
# Caption
'''Relationship between Total Vaccinations and Persons of Last Dose. 
As total vaccinations increase, persons of last dose increase.'''
```

C. Box Plot with Grouping: create a box plot to compare the distribution of a numeric variable among different levels of a factor variable

```python
# Generative AI foundation
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(12, 7))
sns.boxplot(x='WHO_REGION', y='TOTAL_VACCINATIONS_PER100', data=df)
```

```python
# Manual aesthetic improvements
import numpy as np
plt.figure(figsize=(12, 7))
sorted_order = sorted(df['WHO_REGION'].unique())
box_plot = sns.boxplot(x='WHO_REGION', y='TOTAL_VACCINATIONS_PER100', data=df, order=sorted_order)

medians_sorted = df.groupby(['WHO_REGION'])['TOTAL_VACCINATIONS_PER100'].median().sort_index()
median_labels_sorted = [str(np.round(s, 2)) for s in medians_sorted]

pos = range(len(medians_sorted))
for tick, label in zip(pos, box_plot.get_xticklabels()):
    box_plot.text(pos[tick], medians_sorted.iloc[tick] + 5, median_labels_sorted[tick], 
                  horizontalalignment='center', size='x-small', color='black', weight='semibold')

plt.xlabel('WHO Region', fontsize=15)
plt.ylabel('Total Vaccinations per 100 Populations', fontsize=15)
plt.title('Distribution of Total Vaccinations per 100 Populations By WHO Region', fontsize=20)
plt.show()
```

```python
# Alt-text description
'''This plot is about the distribution of the total vaccinations per 100 populations by WHO region. 
The x-axis represents different WHO regions.
The y-axis indicates distribution of the total vaccinations per 100 populations across different WHO regions. 
The figure demonstrates EURO and WPRO have a high median number of total vaccinations per 100 populations and AFRO and EMRO have a low one. 
'''
```

```python
# caption
'''Distribution of Total Vaccinations per 100 Populations By WHO Region. 
EURO and WPRO have a high median and AFRO and EMRO have a low one.'''
```

D. Pairwise Scatter Plots: generate a matrix of scatter plots to visualize pairwise relationships between numerical columns in the dataset. This should result in a grid of scatter plots with each plot showing the relationship between two columns.

```python
# Generative AI foundation 
selected_columns = [
    'TOTAL_VACCINATIONS',
    'PERSONS_VACCINATED_1PLUS_DOSE',
    'PERSONS_LAST_DOSE',
    'PERSONS_BOOSTER_ADD_DOSE',
    'TOTAL_VACCINATIONS_PER100',
    'PERSONS_VACCINATED_1PLUS_DOSE_PER100',
    'PERSONS_LAST_DOSE_PER100'
]

sns.pairplot(df[selected_columns], corner=True,diag_kind="none")

```

```python
# Manual aesthetic improvements
import scipy.stats
selected_columns = [
    'TOTAL_VACCINATIONS',
    'PERSONS_VACCINATED_1PLUS_DOSE',
    'PERSONS_LAST_DOSE',
    'PERSONS_BOOSTER_ADD_DOSE',
    'TOTAL_VACCINATIONS_PER100',
    'PERSONS_VACCINATED_1PLUS_DOSE_PER100',
    'PERSONS_LAST_DOSE_PER100'
]

g = sns.pairplot(df[selected_columns+['WHO_REGION']], corner=True,diag_kind="none",hue='WHO_REGION')            
plt.title('Pairwise Scatter Plots of Selected Numerical Columns',fontsize=20)

for i, col1 in enumerate(selected_columns):
    for j, col2 in enumerate(selected_columns):
        ax = g.axes[i, j]
        if i > j:
            common_index = df[col2].dropna().index.intersection(df[col1].dropna().index)
            x_data = df[col2].loc[common_index]
            y_data = df[col1].loc[common_index]
            ax.scatter(x_data, y_data, s=10, alpha=0.5)
            if not x_data.empty and not y_data.empty:
                r, _ = scipy.stats.pearsonr(x_data, y_data)
                ax.text(0.85, 0.9, f'Ï = {r:.2f}', transform=ax.transAxes, ha='center', va='center', fontsize=10, bbox=dict(facecolor='white', alpha=0.5))
        elif i == j:
            ax.text(0.5, 0.5, col1, transform=ax.transAxes, ha='center', va='center')
            ax.axis('off')
```

```python
# Alt-text description
'''This plot presents a matrix of scatter plots, indicating the relationships between selected numeric variables.
The x-axis and y-axis represent all the selected numeric variables. 
Each individual scatter plot shows the relationship between the respective numeric variables it intersects. 
The figure demonstrates there is a strong positive relationship between the number of persons of at least one dose per 100 populations and the number of persons of last dose (persons who have received all the vaccines to be considered fully 
vaccinated) per 100 populations (p=0.98). 
'''

# caption
'''Pairwise Scatter Plots of Selected Numerical Columns. 
The number of persons of at least one dose per 100 populations is strongly correlated with the number of persons of last dose per 100 populations.'''
```
